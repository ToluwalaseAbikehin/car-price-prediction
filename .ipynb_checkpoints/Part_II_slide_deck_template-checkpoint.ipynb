## (Visualization 2)

> You may not just want to know the car types that are dominating in terms of counts but also in terms of prize as well. Guess the features of the highest prized car: 


```python
def boxgrid(x, y, **kwargs):
    default_color = sb.color_palette()[0]
    sb.boxplot(x, y, color = default_color)
    plt.xticks(rotation = 90)

plt.figure(figsize = [10, 10])
g = sb.PairGrid(data = df, y_vars = ['Price'], x_vars = 'Fuel type',size = 5, aspect = 1.5)
g.map(boxgrid)
plt.xticks(rotation = 90)
plt.title("Boxplot Showing Price and Fuel Type")
plt.show()
;
"""The Diesel Fueled Cars are higher in price than other Fuel types"""
```

```python
plt.figure(figsize = [10, 10])
g = sb.PairGrid(data = df, y_vars = ['Price'], x_vars = 'Color',size = 5, aspect = 1.5)
g.map(boxgrid)
plt.xticks(rotation = 90)
plt.title("Boxplot Showing Price and Color Type")
plt.show();
"""Black and Grey Cars are higher in price than other Color types"""
```

```python
plt.figure(figsize = [10, 10])
g = sb.PairGrid(data = df, y_vars = ['Price'], x_vars = 'Gear box type',size = 5, aspect = 1.5)
g.map(boxgrid)
plt.xticks(rotation = 90)
plt.title("Boxplot Showing Price and Gearbox Type")
plt.show();
"""Tiptronic Cars are higher in price than other Gear box types"""
"""Conclusively, the features that makes the high priced cars include Tiptronic Black Diesel Fueled Cars,followed by automatic Grey diesel fueled cars."""
```
